# Modify a column with a if/else/then logic - np.where()


### SAS code:
>
    data df1;
        set df1;
        format E $4.;
        if id <= 1 then E = 'low';
        else E = 'high';
    run;
    
### Pandas code:    
> 
    df1['E'] = np.where(df1["id"] <= 1, "low", "high")

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(
    {
        "id":[1,1,2,2],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
    )
df1
```

### Method 1 - np.where()

```python
df1['E'] = np.where(df1["id"] <= 1, "low", "high")
df1
```

```python
df1['F'] = np.where(df1["id"] <= 1, df1["A"]+df1["B"] , df1["C"]+df1["D"])
df1
```

### Method 2 - use pandas filter


```python
df1['E'] = "high"
df1.loc[df1["id"] <= 1, 'E'] = "low"
df1
```

```python
df1['F'] = ""
df1.loc[df1["id"] <= 1, 'F'] = df1["A"]+df1["B"]
df1.loc[df1["id"] > 1, 'F'] = df1["C"]+df1["D"]
df1
```
